# Weather Tracker App — Function Demo Notebook 🌦️

This notebook demonstrates the core logic functions of the **Weather Tracker App**, which live in `src/weather_utils.py`.

The Streamlit app (`src/main_app.py`) provides the interactive user interface, while all of the data handling and calculations are performed by the reusable functions shown below. Running them here lets us test the logic directly, without launching the web app.

## 1. Setup — Imports and Path

The utility module lives inside the `src/` folder, so we add that folder to Python's import path before importing it.

````python
import sys
import os

# Add the 'src' folder to the path so we can import our own module
sys.path.append(os.path.join(os.getcwd(), "src"))

import pandas as pd
import weather_utils as utils  # Our custom logic engine

print("Modules imported successfully.")
````

## 2. Load the Weather Database

`load_data()` safely reads `data/weather_observations.csv` into a pandas DataFrame. If the file is missing, it returns an empty DataFrame with the correct columns so the rest of the app never crashes.

````python
# Read every recorded observation from the CSV file
df = utils.load_data()

print(f"Total observations on record: {len(df)}")

# Show the first few rows to preview the structure of the data
df.head()
````

## 3. View Weather Statistics

`get_weather_stats()` calculates the key metrics required by the project: the **average**, **minimum**, and **maximum** temperatures, plus the **most common** weather condition (the mode).

````python
# Calculate the summary statistics across all observations
stats = utils.get_weather_stats()

# Print each metric in a clean, readable format
print(f"Average Temperature : {stats['avg_temp']:.1f} °C")
print(f"Highest Temperature : {stats['max_temp']:.1f} °C")
print(f"Lowest Temperature  : {stats['min_temp']:.1f} °C")
print(f"Most Common Weather : {stats['common_condition']}")
````

## 4. Search Observations by Date

`search_by_date()` filters the database for a specific date string in `MM-DD-YYYY` format and returns all matching rows.

````python
# Look up a known date from the dataset (the July heatwave record)
target_date = "07-15-2025"
results = utils.search_by_date(target_date)

print(f"Found {len(results)} observation(s) for {target_date}:")
results
````

## 5. Stretch Goal 1 — Text-Based Temperature Graph

`generate_text_graph()` builds a simple bar chart out of text characters, where each block represents roughly two degrees Celsius. This gives a quick visual of recent temperature trends without any plotting library.

````python
# Generate a text graph of the 12 most recent observations
graph = utils.generate_text_graph(limit=12)
print(graph)
````

## 6. Stretch Goal 2 — Filter by Month or Season

`filter_by_timeframe()` isolates observations by either a specific month or a grouped Bahrain season (Summer, Winter, or Transitional).

````python
# Example A: every observation recorded in the summer months (Jun-Sep)
summer = utils.filter_by_timeframe("Season", "Summer (Jun-Sep)")
print(f"Summer observations: {len(summer)} rows")
summer.head()
````

````python
# Example B: every observation recorded in a single month (07 = July)
july = utils.filter_by_timeframe("Month", "07")
print(f"July observations: {len(july)} rows")
july.head()
````

## 7. Stretch Goal 3 — Predict Tomorrow's Weather

`predict_tomorrow()` estimates the likely temperature, humidity, and sky condition for a given month, based on the historical records for that month.

````python
# Predict a July outlook ('07') from historical data
prediction = utils.predict_tomorrow("07")

print(f"Predicted Temperature : {prediction['expected_temp']:.1f} °C")
print(f"Predicted Humidity    : {prediction['expected_humidity']:.0f} %")
print(f"Likely Sky Condition  : {prediction['likely_condition']}")
````

## 8. Stretch Goals 4 & 5 — Year-over-Year Comparison & Record Breakers

`get_climate_records()` groups the data by year to build a comparison table, and finds the all-time highest and lowest recorded temperatures.

````python
# Retrieve the year-over-year summary and the record-breaking extremes
records = utils.get_climate_records()

# Goal 5: all-time records
print(f"All-Time Highest: {records['highest_temp']} °C on {records['highest_date']}")
print(f"All-Time Lowest : {records['lowest_temp']} °C on {records['lowest_date']}")
print()

# Goal 4: the year-over-year comparison table
records['yoy_table']
````
